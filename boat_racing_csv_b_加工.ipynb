```python
# 横カラムのデータを縦カラムに加工
df_list = []
columns = ['rece_id','recers_id','age','branch','weight','class_another','nation_win','nation_win2','local_win','local_win2','motor_win2','boat_win2']

df_timetable = pd.DataFrame()
for i in range(len(df)):
  for j in range(1,7):
      rece_id = df.iloc[i,0]
      recers_id = df.loc[i,'{}枠_登録番号'.format(j)]
      age = df.loc[i,'{}枠_年齢'.format(j)]
      branch = df.loc[i,'{}枠_支部'.format(j)]
      weight = df.loc[i,'{}枠_体重'.format(j)]
      class_another = df.loc[i,'{}枠_級別'.format(j)]
      nation_win = df.loc[i,'{}枠_全国勝率'.format(j)]
      nation_win2 = df.loc[i,'{}枠_全国2連対率'.format(j)]
      local_win = df.loc[i,'{}枠_当地勝率'.format(j)]
      local_win2 = df.loc[i,'{}枠_当地2連対率'.format(j)]
      motor_win2 = df.loc[i,'{}枠_モーター2連対率'.format(j)]
      boat_win2 = df.loc[i,'{}枠_ボート2連対率'.format(j)]


      df_list.extend([rece_id,recers_id, age, branch,weight,class_another,nation_win,nation_win2,local_win,local_win2,motor_win2,boat_win2])
      df_before = pd.DataFrame(data =[df_list], columns =columns)
      df_list = []
      df_timetable = pd.concat([df_timetable,df_before],ignore_index=True)

df_timetable = df_timetable.dropna()
df_timetable = df_timetable.reset_index(drop=True)
```

```python
# 前処理
#級別
class_mapping = {'B2':1,'B1':2,'A2':3,'A1':4}
df_timetable['class_another'] = df_timetable['class_another'].map(class_mapping)

#支部
import sklearn
from sklearn import preprocessing
#ラベルエンコード
list_LabelEncode = ['branch']
for label in list_LabelEncode:
    le = sklearn.preprocessing.LabelEncoder()
    le.fit(df_timetable[label])
    df_timetable[label] = le.transform(df_timetable[label])
```
